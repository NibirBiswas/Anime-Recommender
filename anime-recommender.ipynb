```python
# Converting 'sypnopsis' into list
syn['sypnopsis'] = syn['sypnopsis'].str.replace(',', '').str.split().tolist()
syn['Genres'] = syn['Genres'].str.replace(',', '').str.split().tolist()
```

```python
syn['keyword'] = syn['Genres'] + syn['sypnopsis']
```

```python
syn.head()
```

```python
final = syn.drop(columns = ['Genres', 'sypnopsis'])
```

```python
final.head(1)
```

```python
# Convert the 'Score', 'Type', 'Episodes', and 'keyword' columns into a feature matrix
features = final[['keyword']].astype(str).values
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Apply text vectorization to the 'keyword' column using the CountVectorizer
vectorizer = CountVectorizer()
keyword_matrix = vectorizer.fit_transform(features[:, 0])
```

```python
# Calculate the cosine similarity matrix based on the feature matrix
similarity_matrix = cosine_similarity(keyword_matrix)
```

```python
# Define a function to get anime recommendations based on a given anime
def recommend_anime(anime_title, similarity_matrix, data, top_n=5):
    index = data[data['Name'] == anime_title].index[0]
    similarity_scores = list(enumerate(similarity_matrix[index]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    top_anime_indices = [score[0] for score in similarity_scores[1:top_n+1]]
    top_anime_data = data[['Name', 'Score', 'Type', 'Episodes']].iloc[top_anime_indices]
    top_anime_data = top_anime_data.sort_values('Score', ascending=False)
    return top_anime_data
```

```python
# Now we can use the recommend_anime() function to get recommendations.
# Provide the name of an anime and it will return the top recommended anime
```

```python
anime_title = 'Attack on Titan'
recommendations = recommend_anime(anime_title, similarity_matrix, final)
print(recommendations)
```
